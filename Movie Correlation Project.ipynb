```python
#importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the configuration of the plots that we will create
```

```python
#Read in the csv
df = pd.read_csv("movies 2.csv")
```

```python
df['budget'] = df['budget'].astype('int64') 
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

#### The year in the release data and Year column doesn't match so I will make another column with the correct year

```python
df['Year_correct'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
plt.scatter('budget','gross',data=df)
plt.title('Budget vs Gross Earnings')
plt.xlabel("Budget for film")
plt.ylabel("Gross Earnings")
plt.show()
#It seems like a correlation
```

```python
sns.regplot('budget','gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"}) 
```

#### There is a positive correlation between budget and gross. 

```python
df.corr(method='pearson') #only for numeric fields
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation matrix for numeric fields')
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
df_all_numeric = df.copy() #made a copy of the original dataframe

for col_name in df_all_numeric.columns:
    if(df_all_numeric[col_name].dtype=='object'):
        df_all_numeric[col_name] = df_all_numeric[col_name].astype('category')
        df_all_numeric[col_name] = df_all_numeric[col_name].cat.codes
     
df_all_numeric.head()    
```

```python
df.head()
```

```python
correlation_matrix = df_all_numeric.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation matrix for numeric fields')
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show() 
```

```python
df_all_numeric.corr() 
```

```python
correlation_mat = df_all_numeric.corr()

correlation_pairs = correlation_mat.unstack() #Using unstack to get a better idea
correlation_pairs
```

```python
sorted_pairs = correlation_pairs.sort_values()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
```
